```python
import pandas as pd
import os

# 1. Asegurar que la carpeta de destino exista
os.makedirs('../data/processed', exist_ok=True)

# 2. Cargar los datos crudos originales
df_raw = pd.read_csv('../data/raw/responses.csv')

# 3. Seleccionar las variables para el modelo y tratar valores nulos 
# (Imputación con valor neutro 3 en la escala Likert)
columnas_interes = ['Music', 'Slow songs or fast songs', 'Dance', 'Folk', 'Country']
df_limpio = df_raw[columnas_interes].copy()
df_limpio.fillna(3, inplace=True)

# 4. Guardar el dataset procesado en la carpeta requerida
ruta_salida = '../data/processed/dataset_limpio.csv'
df_limpio.to_csv(ruta_salida, index=False)

print(f"¡Dataset procesado y exportado con éxito en: {ruta_salida}!")
```
