```python
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# 1. Cargar datos limpios
df = pd.read_csv('../data/processed/dataset_limpio.csv')

# 2. Escalar
escalador = StandardScaler()
X_escalado = escalador.fit_transform(df)

# 3. Entrenar Modelo 1: K-Means
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')
labels_kmeans = kmeans.fit_predict(X_escalado)
sil_kmeans = silhouette_score(X_escalado, labels_kmeans)

# 4. Entrenar Modelo 2: Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=4)
labels_agg = agg.fit_predict(X_escalado)
sil_agg = silhouette_score(X_escalado, labels_agg)

print(f"Silhouette Score (K-Means): {sil_kmeans:.4f}")
print(f"Silhouette Score (Agglomerative): {sil_agg:.4f}")
print("Conclusión: K-Means obtiene un mejor puntaje de separación, por lo que será el modelo final.")
```
